### Linear Regression on Boston Houses Dataset

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
X = boston.data
y = boston.target
print(X.shape)
print(y.shape)
```

```python
# Normalise this dataset
import numpy as np
u = np.mean(X,axis=0)
std = np.std(X,axis=0)
print(u.shape,std.shape)
```

```python
X = (X-u)/std
pd.DataFrame(X[:5,:]).head()
```
